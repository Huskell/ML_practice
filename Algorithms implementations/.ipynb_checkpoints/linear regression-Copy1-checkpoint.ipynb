```python
import random
import pandas as pd #библиотека для работы с данными
import numpy as np #математическая библиотека

import matplotlib.pyplot as plt # библиотека для визуализации
import time
```

```python
def f(x):
    return (x - 5)**2
```

```python
# рисуем график функции
x = range(-5,15)
y = [f(xn) for xn in x]
plt.plot(x, y)
plt.show()
```

```python
class GradientDescent:
    def __init__(self, step=0.01):
        self.x = random.uniform(-10, 10)
        self.y = self.f(self.x)
        self.step = step
        self.his = {self.x: self.y}

    def history(self, x , y):
        self.his[self.x] = self.y

    def f(self, x):
        return (x - 5)**2

    def f_der(self, x):
        return 2 * (x - 5)

    def gd_step(self):
        self.x = self.x - self.step * (self.f_der(self.x))

    def grad_desc(self, it):
        for i in range(it):
            self.gd_step()
            self.y = self.f(self.x)
            self.history(self.x, self.y)
```

```python
gd = GradientDescent(0.75)
gd.grad_desc(20)
print (gd.x, gd.y)
```

```python
x = range(-5,10)
y = [f(xn) for xn in x]

for i in gd.his.keys():
    plt.plot(x, y)
    plt.plot(i, f(i), 'ro')
    plt.show()
```

```python
from scipy.misc import derivative
```

```python
class GradientDescent_2:
    def __init__(self, step=0.01):
#         self.x = random.uniform(-1, 11)
        self.x = -5
        self.y = f(self.x)
        self.step = step
        self.his = {self.x: self.y}

    def history(self, x , y):
        self.his[self.x] = self.y

    def f(self, x):
        return (x - 5)**2 * np.sin(x)

    def f_der(self, x):
        return derivative(self.f, x)

    def gd_step(self):
        self.x = self.x - self.step * (self.f_der(self.x))

    def grad_desc(self, it):
        for i in range(it):
            self.gd_step()
            self.y = f(self.x)
            self.history(self.x, self.y)
```

```python
gd2 = GradientDescent_2(0.1)
gd2.grad_desc(20)
print (gd2.his)
```
