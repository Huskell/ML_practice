## Линейная регрессия

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', header=None, names=['X', 'Y'])
df.count()
```

```python
def plot_data():
    plt.figure(figsize=(14, 8))
    plt.scatter(df['X'], df['Y'])
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Dataset')

plot_data()
```

```python
def plot_line(w0, w1, color='r'):
    y = w1*df['X'] + w0
    plt.plot(df['X'], y, color, alpha=0.5, label='manifold')
```

```python
plot_data()

plot_line(1, 2, 'r')
plot_line(-40, 2, 'b')
plot_line(5, 1, 'm')
plot_line(-7, 1.5, 'black')
plot_line(7.99, 1.52, 'c')

plot_line(-25, 2, 'y')

plt.show()
```

### Метод наименьших квадратов

```python
class MNK_linear_regressoin:
    def __init__(self):
        self.w1 = 0
        self.w0 = 0
    
    def get_param(self):
        return [self.w0, self.w1]
    
    def fit(self, x, y):
        self.w1 = (np.mean(x*y)-np.mean(y)*np.mean(x))/(np.mean(x**2)-np.mean(x)**2)
        self.w0 = np.mean(y) - self.w1*np.mean(x)
        
    def predict(self, x):
        return self.w1*x + self.w0
```

```python
lr = MNK_linear_regressoin()
print("параметры до обучения на данных: ", lr.get_param())

lr.fit(df['X'],df['Y'])
print("параметры после обучения на данных: ", lr.get_param())

w0_mnk, w1_mnk = lr.get_param()
```

```python
plot_data()

plot_line(*lr.get_param())
plot_line(-25, 2, 'y')
plt.show()
```

### Градиентный спуск

```python
def compute_cost(c, m, x, y):
    total_cost = 0
    N = float(len(x))
    
    # Compute sum of squared errors
    for i in range(0, len(x)):
        xc = x[i]
        yc = y[i]
        total_cost += (yc - (m * xc + c)) ** 2
        
    # Return average of squared error
    return total_cost/N
```

```python
def gradient_descent(x, y, w0_start, w1_start, learning_rate, num_iterations):
    w0 = w0_start
    w1 = w1_start
    cost = []

    for i in range(num_iterations):
        cost.append(compute_cost(w0, w1, x, y))
        w0, w1 = step_gradient(w0, w1, x, y, learning_rate)

    return [w0, w1, cost]

def step_gradient(w0, w1, x, y, learning_rate):
    w0_sum = 0
    w1_sum = 0
    N = float(len(x))

    # Calculate Gradient
    for i in range(0, len(x)):
        xg = x[i]
        yg = y[i]
        w1_sum += - (2/N) * (yg - (w1 * xg + w0)) * xg
        w0_sum += - (2/N) * (yg - (w1 * xg + w0))
    
    # Update current 
    w0_new = w0 - learning_rate * w0_sum
    w1_new = w1 - learning_rate * w1_sum

    # Return updated parameters
    return w0_new, w1_new
```

```python
w0_start = 0
w1_start = 0
learning_rate = 0.00001
num_iterations = 100

w0_gd, w1_gd, cost = gradient_descent(df['X'], df['Y'], w0_start, w1_start, learning_rate, num_iterations)

# Print optimized parameters
print ('w0 = :', w0_gd)
print ('w1 = :', w1_gd)

# Print error with optimized parameters
print ('Q = :', compute_cost(w0_gd, w1_gd, df['X'], df['Y']))
```

```python
plt.plot(cost)
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.title('Cost per iteration')
plt.grid(True)
plt.show()
```
