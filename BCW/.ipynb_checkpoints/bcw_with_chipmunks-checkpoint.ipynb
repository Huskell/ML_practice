## 3 step

Гистограммы распределения признаков

```python
df['diagnosis'].hist()
```

```python
# гистограмма
ax = sns.countplot(df['diagnosis'],label="Count")       # M = 212, B = 357
B, M = df['diagnosis'].value_counts()
print('Number of Benign: ',B)
print('Number of Malignant : ',M)
```

Описание в df.describe() говорит что разброс значений между признаками большой. Перед визуализацией было бы неплохо их нормализовать или стандартизировать

Стандартизация данных — это получение датасета с нулевым значением среднего арифметического и стандартным отклонением, которое равно единице.

```python
x = df.drop(['diagnosis'], axis=1)
x_std = (x - x.mean())/x.std()
```

Нормализация предполагает замену номинальных признаков так, чтобы каждый из них лежал в диапазоне от 0 до 1

```python
x_norm = (x - x.min())/(x.max() - x.min())
```

```python
# переменная содержит ответы
y = df['diagnosis']
```

```python
df_std
```

```python
df_std = pd.concat([y, x_std.iloc[:, 0:10]], axis=1)
df_std = pd.melt(df_std, id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(16, 10))
sns.boxplot(x='features', y='value', hue="diagnosis", data=df_std)
plt.xticks(rotation=90)
```

```python
df_std = pd.concat([y, x.iloc[:, 0:1]], axis=1)
df_std = pd.melt(df_std, id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(8, 8))
sns.boxplot(x='features', y='value', hue="diagnosis", data=df_std)
plt.xticks(rotation=90)
```

Медиа́на (от лат. mediāna «середина») набора чисел — число, которое находится в середине этого набора, если его упорядочить по возрастанию.

1: на самлм левом графике мы видим что по данному признаку типы ответов хорошо отделимы. Медианы расположены далеко друг от друга.
на самом правом графике обратная сиьуация и ответы практически невозможно отделить друг от друга из-за близости значений в признаке.

Предположение: правые признаки скорее всего слабые и будут вносить шум в наше решение

2: смотря на признаки 3 и 4 (7 и 8), мы видим что медианы практически совпадаю. Так же похожи размеры коробок. Можно предположить что эти признаки имеют сильную корреляцию

```python
df_std = pd.concat([y, x_norm.iloc[:, 0:10]], axis=1)
df_std = pd.melt(df_std, id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(16, 10))
sns.boxplot(x='features', y='value', hue="diagnosis", data=df_std)
plt.xticks(rotation=90)
```

При стандартизации мы центрируем ВСЕ признаки относительно 0. Удобно для сравнени япризнаков между собой

При нормализации мы центрируем КАЖДЫЙ признак относительно среднего значения этого признака

### Лучше поссать, чем зассать, а тем более сосать

```python
sns.jointplot(x.loc[:,'perimeter_mean'], x.loc[:,'area_mean'], height=8, kind="reg", color="blue")
x[['perimeter_mean', 'area_mean']].corr() #pearson's correlation
```

```python
g = sns.jointplot(x.loc[:,'radius_mean'], x.loc[:,'concave points_mean'], height=8, kind="reg", color="blue")
x[['radius_mean', 'concave points_mean']].corr() #pearson's correlation
```

```python
df_pg = x.loc[:,['radius_worst','perimeter_worst','area_worst']]
g = sns.PairGrid(df_pg, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```
